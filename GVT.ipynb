 #  Global Virtual Team (GVT) Performance Development Ontology in Global Software Development (GSD)--(GVTPDOnto)
This project aims to build an ontology of the Global Virtual Team (GVT) performance development model in Global Software Development (GSD)(GVTPDModel), for unifying knowledge related to the factors affecting GVT performance in GSD and the concepts and goals associated with the GSD that represent the entities of the GVTPDModel, and to facilitate access to this knowledge and its dissemination among workers in this field. 

# Creat and save ontology

```python
from owlready2 import *
import ontor
import webbrowser
import os
def select_drive(x):# Functions to select drive to save ontology in.
    global drive   
    if "D" in x or "d" in x:
        drive="D:\\ "      
    else: 
        if "E" in x or "e" in x:
            drive="E:\\ " 
        else:
            if "F" in x or "f" in x:
                drive="F:\\ "
            else:
                print("Wrong Input!")
                query = input("Enter the correct drive: D , E or F:")
                select_drive (query)
prefix= input("Enter ontology name to save: or press C to canceled ")
if "C" in prefix or "c" in prefix:      
    print("save file canceled")
else:    
    query = input("Which drive do you want to save in ? D , E or F:")
    select_drive(query)
    onto = get_ontology("http://www.semanticweb.org/dell/ontologies/2023/3/"+prefix)
    filePath=drive+prefix+".owl"
    onto.save(filePath, format = "rdfxml")
    print("Ontology saved in:"+drive+prefix+".owl")
namespac=onto.get_namespace("http://www.semanticweb.org/dell/ontologies/2023/3/"+prefix)
```
